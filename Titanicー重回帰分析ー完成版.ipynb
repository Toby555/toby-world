ーEmbarked を、S：１、Q:２、C:３に変換するー

ー Mr.、Mrs. などの名称別に、年齢の平均値補完を行うー

```python
mr_lst, miss_lst, mrs_lst, master_lst = [],[],[],[]

name_lst= []
```

```python
for name, age in zip (train.Name , train.Age):
    if age>0:
        if "Mr." in name:
            mr_lst.append(age)
            
        elif "Miss." in name:
            miss_lst.append(age)
            
        elif "Mrs." in name:
            mrs_lst.append(age)
            
        elif "Master" in name:
            master_lst.append(age)
            
        else:
            name_lst.append(name)
```

```python
print("Mr    の平均値は{}".format(round(sum(mr_lst)/len(mr_lst),1)))
print("Mrs   の平均値は{}".format(round(sum(mrs_lst)/len(mrs_lst),1)))
print("Miss  の平均値は{}".format(round(sum(miss_lst)/len(miss_lst),1)))
print("Masterの平均値は{}".format(round(sum(master_lst)/len(master_lst),1)))
```

```python
age = DF_train["Age"].fillna(-1)
age_1 = list(age.values.flatten())
name = DF_train["Name"]
```

```python
for i in range(0,len(age_1)):
    if "Miss" in name[i]:
        if age[i] == -1:
            age[i] = 21.8
    
    if "Mr" in name[i]:
        if age[i] == -1:
            age[i] = 32.4
    
    if "Mrs" in name[i]:
        if age[i] == -1:
            age[i] = 35.9
    
    if "Master" in name[i]:
        if age[i] == -1:
            age[i] = 4.6
```

ー重回帰分析で予測するー

```python
# 重回帰の線形回帰を実行
from sklearn.linear_model import LinearRegression

model= LinearRegression()
model.fit(X_train, Y_train)
```

ーグリッドサーチで最適な値を求めるー

```python
from sklearn.metrics import accuracy_score
```

```python
lst=[]

for i in np.arange(0.35, 0.65, 0.01):
    Y_train_pred = model.predict(X_train)
    Y_train_pred = np.where(Y_train_pred > i, 1, 0)
    print("{:.2f}の時のスコアは{:.4f}".format(i,accuracy_score(Y_train,Y_train_pred)))
    
    lst.append(round(accuracy_score(Y_train,Y_train_pred),4))

print("Maxの時は i={}で、スコアは{:.4f}".format(0.35+0.01*int(lst.index(max(lst))),max(lst)))
```
